```python
import numpy as np
import matplotlib.pyplot as plt
```

# Cisco Catalyst 9120AX Simulation
### Specifications:
* Operates at 2.4GHz or 5.0GHz
* 2.4GHz transmit power can either be 23dBm (200mW) or -4dBm
* 2.4GHz peak gain is 4dBi, omnidirectional in azimuth
### Assumptions:
* Will assume 0 dBi net gain, for purposes of this simulation
* Will assume 0 dBi reciever

### Frii equation:


```python
speedOfLight = 3e8  # m/s
frequency = 2.4e9  # Hz
waveLength = speedOfLight / frequency  # m
transmitPowerMilliWatts = 200  # mW
transmitGain = 1
receiveGain = 1

distance = np.linspace(0.5, 10, 50)
receivePowerMilliWatts = transmitPowerMilliWatts * transmitGain * receiveGain * (waveLength/(4*np.pi*distance))**2

plt.plot(distance, receivePowerMilliWatts)
plt.title("Power vs Distance from 23dBm Sourc at 2.4 GHz")
plt.xlabel("Distance (m)")
plt.ylabel("Power (mW)")
```

```python
plt.plot(distance, 10*np.log10(receivePowerMilliWatts))
plt.title("Power vs Distance from 23dBm Source at 2.4 GHz")
plt.xlabel("Distance (m)")
plt.ylabel("Power (dBm)")
```

# Dedicated RF Energy Harvesting using Powercast P2110B

Datasheet: https://www.powercastco.com/wp-content/uploads/2021/06/P2110B-Datasheet-Rev-3.pdf

### Specifications
* Operates at 915 MHz
* Minimum operating power input is -12 dBm
### Assumptions:
* Will assume 0 dBi net gain
* Will assume 0 dBi reciever


```python
speedOfLight = 3e8  # m/s
frequency = 915e6  # Hz
waveLength = speedOfLight / frequency  # m
transmitPowerMilliWatts = 4000  # mW
transmitGain = 1
receiveGain = 1

distance = np.linspace(0.1, 7, 50)
receivePowerMilliWatts = transmitPowerMilliWatts * transmitGain * receiveGain * (waveLength/(4*np.pi*distance))**2

plt.plot(distance, receivePowerMilliWatts)
plt.title("Power vs Distance from 23dBm Source at 915 MHz")
plt.xlabel("Distance (m)")
plt.ylabel("Power (mW)")
```

```python
plt.plot(distance, 10*np.log10(receivePowerMilliWatts))
plt.hlines(-12, distance[0], distance[-1], linestyles='dashed')
plt.title("Power vs Distance from 23dBm Source at 915 MHz")
plt.xlabel("Distance (m)")
plt.ylabel("Power (dBm)")
```
