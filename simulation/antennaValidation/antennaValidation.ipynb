```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = 3e8  # Speed of light in m/s
f = 915e6  # Frequency in Hz

freeSpaceWavelength = c / f

transmitPowerDBM = 23  # dBm

transmitPowerMilliWatts = 10**(transmitPowerDBM/10)  # mW
transmitGainDBI = 0
receiveGainDBI = 0
transmitGain = 10**(transmitGainDBI/10)
receiveGain = 10**(receiveGainDBI/10)

distance = np.linspace(0.05, 1, 80)

receivePowerDBM = transmitPowerDBM + transmitGainDBI + receiveGainDBI + 20*np.log10(freeSpaceWavelength/(4*np.pi*distance))

measuredDistance = np.arange(5, 105, 5)/100
measuredReceivedPowerDBM = np.array([12.8,
10.2,
8,
6.7,
5.5,
3,
1.1,
0.3,
-1.3,
-2.7,
-2.4,
-6.3,
-8.7,
-7,
-6.7,
-6.7,
-4,
-8,
-9,
-8.2])

plt.figure(figsize=(10, 5), dpi=500)
plt.plot(distance, receivePowerDBM, label="Simulated Power")
plt.plot(measuredDistance, measuredReceivedPowerDBM, label="Measured Power")
plt.title(f"Power vs Distance from {transmitPowerDBM}dBm Source at 915 MHz")
plt.xlabel("Distance (m)")
plt.ylabel("Power Received (dBm)")
plt.grid("True")
plt.legend()


```
