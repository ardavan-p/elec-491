```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = 3e8  # Speed of light in m/s
f = 915e6  # Frequency in Hz

wavelength = c / f

eirp = 36  # dBm
tx_gain = 6  # dBi
rx_gain = 0  # dBi

tx_power = eirp - tx_gain  # dBm

tire_path_loss = 10  # dB


d = np.linspace(0.05, 2, 400)

rx_power = eirp + rx_gain + 20*np.log10(wavelength/(4*np.pi*d)) - tire_path_loss

plt.figure(figsize=(10, 5), dpi=500)
plt.plot(d, rx_power, label="Simulated Power")
plt.title(f"Power vs Distance from {tx_power}dBm RF Generator at 915MHz")
plt.xlabel("Distance (m)")
plt.ylabel("Power Received (dBm)")
plt.grid("True")
plt.legend()
```
