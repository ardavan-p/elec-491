```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = 3e8  # Speed of light in m/s
f = 2.4e9  # Frequency in Hz

wavelength = c / f

tx_power_dbm = 0  # dBm

tx_gain_dbi = 0
rx_gain_dbi = 0

d = np.linspace(0.05, 10, 400)

rx_power_dbm = tx_power_dbm + tx_gain_dbi + rx_gain_dbi + 20*np.log10(wavelength/(4*np.pi*d))

plt.figure(figsize=(10, 5), dpi=500)
plt.plot(d, rx_power_dbm, label="Simulated Power")
plt.title(f"Power vs Distance from {tx_power_dbm}dBm Source at 2.4GHz")
plt.xlabel("Distance (m)")
plt.ylabel("Power Received (dBm)")
plt.grid("True")
plt.legend()


```
